# COORDINATES AND DIAGONALS

```python
#Controlling Axes and Adding Diagonals
g = sns.FacetGrid(movies, row = 'Genre', col = 'Year', hue = 'Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
g = g.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
#g.set(xlim=(0,100), ylim=(0,100))
for ax in g.axes.flat: #g. axes (taking arguments from facet grid) . flat(making them like a list)
    ax.plot((0,100),(0,100), c='gray', ls='--')
#if dot is above the diagonal it means that audience gave better rating than critics
g.add_legend()
```

# BUILDING DASHBOARDS IN PYTHON

```python
sns.set_style("darkgrid")
f, axes = plt.subplots(2, 2, figsize=(10, 10))
k1 = sns.kdeplot(x = movies.BudgetMillions, y = movies.AudienceRating, ax=axes[0, 0]) 
k2 = sns.kdeplot(x = movies.BudgetMillions, y = movies.CriticRating, ax=axes[0, 1])
k3 = sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year', y='CriticRating', ax=axes[1,0])
k4 = sns.kdeplot(x= movies.CriticRating,y= movies.AudienceRating, shade = True, cmap='Reds', ax=axes[1,1])
#axes[1,1].hist(movies.CriticRating, bins=15) #for non seaborn plots we need to specify it in different way
#axes is a pyplot object and thats why it wouldnt work for sns plots
#TIP:
k1.set(xlim=(-40,250))
k2.set(xlim=(-40,250))

plt.show()
```

# STYLING TIPS

```python
sns.set_style("whitegrid", {"axes.facecolor":"black"}) #white, whitegrid, darkgrid, ticks
f, axes = plt.subplots(2, 2, figsize=(10, 10))
#plot 0,0
k1 = sns.kdeplot(x = movies.BudgetMillions, y = movies.AudienceRating, ax=axes[0, 0], shade=True,thresh=0, cmap='inferno') 
k1b = sns.kdeplot(x = movies.BudgetMillions, y = movies.AudienceRating, ax=axes[0, 0], cmap='cool') #contour

#plot 0,1
k2 = sns.kdeplot(x = movies.BudgetMillions, y = movies.CriticRating, ax=axes[0, 1],thresh=0 ,shade=True, cmap='inferno')
k2 = sns.kdeplot(x = movies.BudgetMillions, y = movies.CriticRating, ax=axes[0, 1], cmap='cool')#contour

#plot 1,0
k3 = sns.violinplot(data=movies, x='Year', y='BudgetMillions', palette='YlOrRd', ax=axes[1,0])

#plot 1,1
k4 = sns.kdeplot(x= movies.CriticRating,y= movies.AudienceRating, shade = True, cmap='Blues_r' ,ax=axes[1,1])
k4 = sns.kdeplot(x= movies.CriticRating,y= movies.AudienceRating, cmap='gist_gray_r', ax=axes[1,1])#contour


k1.set(xlim=(-40,250))
k2.set(xlim=(-40,250))

plt.show()
```

# FINISHING TOUCHES

```python
genre_list = list(movies.Genre.cat.categories)
plot_list = []
labels_list = []
for genre in genre_list:
        plot_list.append(movies[movies.Genre==genre].BudgetMillions)
        labels_list.append(genre)

sns.set_style("whitegrid")

fig, ax = plt.subplots()      #create subplot with 1 item to increase size of plot fast  
fig.set_size_inches(11.7, 8.27) #size of A4 paper
plt.hist(plot_list, bins=30, stacked=True, label= labels_list) #for colors type: color = [list of colors]
plt.title("Movies Budget Distribution", fontsize=35, color="DarkBlue", fontname="Arial")
plt.ylabel("Number of Movies", fontsize=35, color="Red", fontname="Arial")
plt.xlabel("Budget", fontsize=35, color="Green", fontname="Arial")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(fancybox=True, shadow=True,framealpha=1, prop={'size':20})
plt.show()
```
